### What are Word Embeddings?

Word embeddings are numerical representations of words that capture their semantic meaning and relationships with other words. In these representations, words with similar meanings are located closer to each other in a multi-dimensional vector space.

Key characteristics and benefits:

*   **Distributed Representation**: Unlike one-hot encoding where each word is an independent vector, word embeddings represent words as dense vectors where each dimension captures some feature of the word's meaning.
*   **Semantic Relationships**: They can capture relationships like synonyms, antonyms, and even analogies (e.g., "king - man + woman = queen").
*   **Dimensionality Reduction**: They reduce the high dimensionality of sparse representations (like one-hot vectors) to a much lower-dimensional dense vector.
*   **Input for Machine Learning**: Word embeddings are widely used as input features for various Natural Language Processing (NLP) tasks, including:
    *   Machine Translation
    *   Sentiment Analysis
    *   Text Classification
    *   Question Answering
    *   Information Retrieval

**Common Techniques for Generating Word Embeddings:**

1.  **Word2Vec**: Popularized by Google, it includes two main architectures:
    *   **Skip-gram**: Predicts context words given a target word.
    *   **CBOW (Continuous Bag of Words)**: Predicts a target word given its context words.
2.  **GloVe (Global Vectors for Word Representation)**: Combines global matrix factorization and local context window methods.
3.  **FastText**: An extension of Word2Vec that represents words as n-grams of characters, allowing it to handle out-of-vocabulary words and morphologically rich languages more effectively.
4.  **BERT, ELMo, GPT (Contextual Embeddings)**: More recent models that generate embeddings that are sensitive to the context in which a word appears, providing different embeddings for the same word based on its usage in a sentence.

### 1. Prepare Text Data

Word2Vec models require a corpus of text. For simplicity, we'll use a small list of sentences. In a real-world scenario, this would be a much larger dataset. We also need to tokenize the sentences (split them into words).

```python
import nltk
from nltk.tokenize import word_tokenize
import gensim
from gensim.models import Word2Vec

# Download Punkt tokenizer for nltk if not already present
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

# Download punkt_tab as well, if needed by word_tokenize
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

# Sample corpus
corpus = [
    "I love natural language processing",
    "Word embeddings are fascinating",
    "Gensim is a powerful library for NLP",
    "Machine learning is a subset of AI",
    "Deep learning revolutionized many fields",
    "NLP tasks include sentiment analysis and machine translation",
    "Embeddings capture semantic relationships between words",
    "The quick brown fox jumps over the lazy dog"
]

# Tokenize the corpus
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

print("Tokenized Corpus:")
for sentence in tokenized_corpus:
    print(sentence)
```

### 2. Train the Word2Vec Model

Now we'll train the Word2Vec model using our tokenized corpus. Key parameters include:
*   `vector_size`: Dimensionality of the word vectors.
*   `window`: Maximum distance between the current and predicted word within a sentence.
*   `min_count`: Ignores all words with total frequency lower than this.
*   `sg`: Training algorithm (0 for CBOW, 1 for Skip-gram).

```python
# Train the Word2Vec model using CBOW (sg=0)
model_cbow = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, workers=4, sg=0)

# Train the Word2Vec model using Skip-gram (sg=1)
model_skipgram = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1, workers=4, sg=1)

print("Word2Vec CBOW model trained successfully as 'model_cbow'!")
print("Word2Vec Skip-gram model trained successfully as 'model_skipgram'!")
```

### 3. Explore Word Embeddings

Now that the model is trained, we can explore the learned word embeddings. We can:
*   Get the vector for a specific word.
*   Find words most similar to a given word.
*   Perform word analogies.

```python
# Get the vector for a word
for model_name, model in [('CBOW', model_cbow), ('Skip-gram', model_skipgram)]:
    print(f"\n--- Exploring {model_name} Model ---")
    try:
        # Check if the word 'language' exists in the vocabulary before accessing
        if 'language' in model.wv:
            word_vector = model.wv['language']
            print(f"Vector for 'language' (first 10 dimensions) from {model_name}:")
            print(f"{word_vector[:10]}\n")
        else:
            print(f"'language' not in {model_name} model vocabulary.\n")

        # Find most similar words
        if 'processing' in model.wv:
            similar_words = model.wv.most_similar('processing', topn=5)
            print(f"Words similar to 'processing' from {model_name}:\n{similar_words}\n")
        else:
            print(f"'processing' not in {model_name} model vocabulary.\n")

        if 'ai' in model.wv:
            similar_words_ai = model.wv.most_similar('ai', topn=3)
            print(f"Words similar to 'ai' from {model_name}:\n{similar_words_ai}\n")
        else:
            print(f"'ai' not in {model_name} model vocabulary.\n")

        # Word analogies (e.g., 'king' - 'man' + 'woman' = 'queen')
        # This works better with a larger vocabulary, but we can try with our small corpus
        # Example: what is to 'learning' as 'natural' is to 'language'?

        # Note: With a very small corpus, results for analogies might not be very meaningful
        # as the model hasn't seen enough patterns.

        try:
            analogy_result = model.wv.most_similar(positive=['nlp', 'machine'], negative=['language'], topn=1)
            print(f"'nlp' - 'language' + 'machine' = {analogy_result} from {model_name}\n")
        except KeyError as e:
            print(f"Could not perform analogy for {model_name} due to missing word: {e}. \nThis often happens with small vocabularies.\n")

        try:
            analogy_result = model.wv.most_similar(positive=['deep', 'machine'], negative=['learning'], topn=1)
            print(f"'deep' - 'learning' + 'machine' = {analogy_result} from {model_name}\n")
        except KeyError as e:
            print(f"Could not perform analogy for {model_name} due to missing word: {e}. \nThis often happens with small vocabularies.\n")

    except Exception as e:
        print(f"An error occurred while exploring {model_name} model: {e}")
```

```python
print(f"\nVocabulary size of CBOW model: {len(model_cbow.wv.key_to_index)}")
print(f"Vocabulary size of Skip-gram model: {len(model_skipgram.wv.key_to_index)}")
```

**Observation:**
1. Word2Vec successfully converted words into dense vector representations.
2. Semantic similarity was observed between related terms like "AI" and "machine".
3. Some irrelevant similarities occurred due to the small dataset.
4. Word arithmetic operations demonstrated the ability of embeddings to capture relationships.
5. CBOW and Skip-gram produced nearly identical outputs due to limited data.
