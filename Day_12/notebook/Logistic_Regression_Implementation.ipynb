```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
#dataset
data = pd.read_csv("../data/insurance_data.csv")
print(data.head())
```

```python
X = data[['age']] # Independent
Y = data['bought_insurance'] # Dependent (0 or 1)
```

```python
# train test split
X_train, X_test, Y_train, Y_test = train_test_split(
X, Y, test_size=0.2, random_state=42
)
```

```python
#Tarin Model
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#Predict
Y_pred = model.predict(X_test)
```

```python
#Confusion Matrix and Accuracy
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("Accuracy:", accuracy_score(Y_test, Y_pred))
```

##### Visual Representation

```python
# Scatter plot (actual data)
plt.scatter(X, Y, color='blue', label='Actual Data')
# Sort values for smooth curve
X_sorted = np.sort(X.values, axis=0)
# Predict probabilities
Y_prob = model.predict_proba(X_sorted)[:, 1]
# Plot sigmoid curve
plt.plot(X_sorted, Y_prob, color='red', linewidth=3, label='Sigmoid Curve')
# Decision boundary (0.5)
plt.axhline(y=0.5, color='yellow', linestyle='--', label='Decision Boundary')
# Labels & title
plt.xlabel("Age")
plt.ylabel("Probability / Class")
plt.title("Logistic Regression - Insurance Dataset")
# Legend
plt.legend()
plt.grid()
plt.show()
```

#### Observations:
* The scatter plot (blue points) represents the actual data distribution of age vs insurance purchase.
* The red curve represents the sigmoid function, which models the probability of buying insurance.
* The sigmoid curve shows a smooth S-shaped transition from 0 to 1.
* For lower age values, the probability of buying insurance is close to 0.
* As age increases, the probability gradually increases.
* For higher age values, the probability approaches 1, indicating a higher likelihood of buying insurance.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy:", accuracy_score(Y_test, Y_pred))
print("Precision:", precision_score(Y_test, Y_pred))
print("Recall:", recall_score(Y_test, Y_pred))
print("F1 Score:", f1_score(Y_test, Y_pred))
```
