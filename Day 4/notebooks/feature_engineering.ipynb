```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

## Load Data

```python
df = pd.read_csv("../data/salary_data.csv")
```

```python
print("🔍 Original Data:\n", df.head()) #first 5 values
```

```python
df.shape #Shape(rows,Columns)
```

## Checking For Missing values

```python
df.isnull().sum()
```

## Handle Missing Values

```python
# Fill Age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df['Age'] = df['Age'].round(0).astype(int) # Round of the Age
df
```

```python
# Fill Salary with median (better for large values)
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
df
```

```python
#Remove duplicates
df = df.drop_duplicates()
df
```

```python
df.shape # No Duplicates
```

##  Statistical Summary

```python
df.describe()
```

###  Feature Scaling & Normalization

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=['Age', 'Salary'])
```

```python
scaled_data
```

```python
import os

os.makedirs("outputs", exist_ok=True)
```

```python
# Save files tto outputs folder
df.to_csv("outputs/cleaned_data.csv", index=False)
scaled_df.to_csv("outputs/scaled_data.csv", index=False)
```

## What I Undersatnd ...

* Scaling converts different values (like Age = 22, Salary = 100000) into a common range (0 to 1) so they can be compared properly.
* Without scaling, large values (like Salary) dominate small values (like Age).
* Scaling ensures every feature gets equal importance.
