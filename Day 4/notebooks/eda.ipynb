# **SECTION 3: EXPLORATORY DATA ANALYSIS (Visualization)**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### **Load Datasets**

```python
df = pd.read_csv("../data/eda_dataset.csv")
```

```python
df
```

#### **Correlation Map**

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

##### Observation:

* All features (Age, Salary, Experience, Score) show strong positive correlation
* Experience and Salary have the highest correlation → strongest relationship
* Age and Experience are highly correlated → logical real-world relation
* Age and Salary also show strong positive correlation
* Score is positively related to all features → increases with better profile
* No negative correlation found in the dataset
* Correlation values are close to +1, indicating strong relationships
* Heatmap color (red/warm tones) confirms high positive correlation

```python
sns.pairplot(df)
plt.show()
```

##### Observation:

* All feature pairs show positive relationships (upward trend in scatter plots)
* Experience vs Salary shows the strongest linear relationship → very clear upward line
* Age vs Experience is highly correlated → older people tend to have more experience
* Age vs Salary also shows strong positive correlation
* Score increases with Age, Salary, and Experience → better overall profile leads to higher score
* No negative trends observed in any pair → dataset is consistent
* Scatter plots are well-clustered and linear → good for ML models
* Diagonal plots (histograms) show smooth distribution of each feature

```python
sns.scatterplot(x='Experience', y='Salary', data=df)
plt.show()
```

##### Observation:

As the Experience increases Salary Increases

```python
sns.boxplot(data=df)
plt.show()
```

##### Observation:

* Most data values lie within a compact range, showing consistency in the dataset
* Median (middle line) is clearly visible → represents the central value of each feature
* No extreme outliers observed → dataset is clean and well-balanced
* Salary shows a wider spread compared to other features → higher variation
* Age has a relatively smaller spread → values are closely grouped
* Score values are mostly concentrated in the higher range → good performance overall

```python
df.hist(figsize=(10,8))
plt.show()
```

##### Observation:

* Fairly uniform distributions across Age, Salary, and Experience → no strong skew or outliers
* Salary and Experience show gradual spread → consistent progression pattern (no sudden jumps)
* Score slightly clustered toward higher values (85–95) → mild positive skew / high-performance bias
* Overall data looks synthetic/controlled → too clean and evenly distributed for real-world data
