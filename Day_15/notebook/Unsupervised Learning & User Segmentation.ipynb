## **Elbow Method**

_In K-Means clustering, the algorithm partitions data into k clusters by minimizing the distances between points and their cluster centroids. However, deciding the ideal k is not straightforward. The Elbow Method helps by plotting the Within-Cluster Sum of Squares (WCSS) against increasing k values and looking for a point where the improvement slows down, this point is called the "elbow."_

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate data (if not already defined)
X, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=0)
```

```python
# Create models
k1 = KMeans(n_clusters=5, init='random', random_state=42).fit(X)
k2 = KMeans(n_clusters=5, init='random', random_state=10).fit(X)
k3 = KMeans(n_clusters=5, init='k-means++', random_state=42).fit(X)

models = [
    ("Random Init (seed=42)", k1),
    ("Random Init (seed=10)", k2),
    ("K-Means++ Init", k3)
]
```

```python
# Fixed colors
colors = ['red', 'blue', 'pink', 'orange', 'purple']

plt.figure(figsize=(18, 5))

for i, (title, model) in enumerate(models):
    y_pred = model.predict(X)

    plt.subplot(1, 3, i+1)

    # Plot each cluster with fixed color
    for j in range(5):
        plt.scatter(
            X[y_pred == j, 0],
            X[y_pred == j, 1],
            s=50,
            c=colors[j],
            label=f'C{j+1}'
        )

    # Plot centroids
    plt.scatter(
        model.cluster_centers_[:, 0],
        model.cluster_centers_[:, 1],
        s=200,
        c='green',
        marker='X',
        label='Centroids'
    )

    plt.title(title)
    plt.legend()

plt.tight_layout()
plt.show()
```

**Random Init (seed = 42):**
* Clusters are reasonably formed, but not perfectly centered.
* Centroids depend on initial random positions.
* May still converge to a good but not optimal solution.
* Slight overlap or imbalance can appear.

**Random Init (seed = 10):**
* Different clustering result despite same data.
* Some clusters look shifted or misgrouped.
* Demonstrates high sensitivity to initialization.
* Can converge to a worse local minimum.

**K-Means++ Init**
* The first centroid is chosen randomly.
* The remaining centroids are selected such that they are far away from each other.
* This ensures better initial separation of clusters.
* Clusters are more accurate and well-separated
* The algorithm converges faster
* Results are more stable and consistent

**Observations**
* K-Means is initialization-dependent
* Random initialization → inconsistent results
* K-Means++ → better starting points → better clusters
* Same dataset + different init = different outputs
* K-Means finds local minimum, not guaranteed global minimum

```python

from sklearn.preprocessing import StandardScaler

# Features
X = df[['cgpa', 'iq']]

# IMPORTANT: Scaling (very important in KMeans)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# Plot clusters
plt.figure(figsize=(8,6))

colors = ['red', 'blue', 'green','pink']

for i in range(4):
    plt.scatter(
        X_scaled[y_kmeans == i, 0],
        X_scaled[y_kmeans == i, 1],
        s=100,
        c=colors[i],
        label=f'Cluster {i+1}'
    )

# Plot centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=300,
    c='yellow',
    marker='X',
    label='Centroids'
)

plt.title('Student Clusters (CGPA vs IQ)')
plt.xlabel('CGPA (scaled)')
plt.ylabel('IQ (scaled)')
plt.legend()
plt.grid(True)
plt.show()
```

* The cluster centers obtained using random initialization (seed = 42 and seed = 10) are different from each other, showing variability in results.
* This happens because:
     * Initial centroids are chosen randomly
     * The algorithm may converge to different local minima
* The cluster centers obtained using k-means++ initialization are:
     * More stable and consistent
     * Better aligned with the actual data distribution
* Therefore:
Random initialization → unstable results
