## 1. Array Creation Functions


```python
import numpy as np
arr = np.array([1, 2, 3]) #NumPy array me convert
```

```python
np.zeros((2, 3)) #2x3 matrix of 0
```

```python
np.ones((2, 2)) #2x3 matrix of 1
```

```python
np.arange(0, 10, 2) #createbthe list of element from 0 to 8 having difference 2
```

```python
np.linspace(0, 1, 5) #Equal interval values generate
```


## 2. Reshaping & Structure


## 3. Mathematical Operations (Vectorization)


```python
arr = np.array([10, 20, 30])

print(arr + 5)
print(arr * 2)
```

```python
np.sum(arr) #Sum of all Elements
```

```python
np.mean(arr) #Mean
```


## 4. Indexing & Slicing


```python
arr = np.array([10, 20, 30, 40])

print(arr[0])     # First element
print(arr[1:3])   # Slice
```


## 5. Matrix Operations


```python
a = np.array([[1,2],
              [3,4]])
b = np.array([[5,6],
              [7,8]])

print(a + b) #Addition
```

```python
print(a.T) #Transpose of matrix
```

```python
np.dot(a, b) #Matrix Multiplication
```

## What is Vectorization?

* Vectorization means performing operations on entire datasets at once instead of using loops.
* It allows speedup the calculation.
* In normal Python, loops process data element by element, which is slower.
* NumPy enables vectorization by applying operations directly on arrays.
* It is widely used in AI/ML for handling large datasets efficiently.
* Vectorization improves performance, reduces code complexity, and makes programs more readable.
