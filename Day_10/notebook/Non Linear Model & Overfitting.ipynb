```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
# Generate same data
np.random.seed(42)
X = 6 * np.random.rand(100, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(100, 1)
```

```python
# Smooth input for plotting
X_new = np.linspace(-3, 3, 200).reshape(-1, 1)
```

```python
# Different depths
depths = [2, 5, 20]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, label="Actual Data")
```

```python
for d in depths:
    tree = DecisionTreeRegressor(max_depth=d)
    tree.fit(X, y)
    y_pred = tree.predict(X_new)

    plt.plot(X_new, y_pred, label=f"max_depth={d}")

plt.title("Effect of max_depth on Decision Tree")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

**Reflection: "Why is a 'jittery' model that hits every training
point actually worse for making future predictions than a smooth 'curvy' model?"**

### Reflection:

* A jittery model that hits every training point is worse for future predictions because it memorizes the noise in the training data instead of learning the true underlying pattern.

* When the model is too complex (high max_depth), it creates many small splits and produces irregular, step-like predictions that perfectly match the training data. However, this leads to poor generalization, meaning the model performs badly on new, unseen data.

* In contrast, a smoother model captures the overall trend of the data without being influenced by noise. This allows it to generalize better and make more reliable predictions on future data.

* Therefore, a balance between model complexity and simplicity is essential to achieve good performance.
