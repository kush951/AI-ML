# Understanding Mathematical Intuition Behind Rigid Regression

#### Ridge regression:
##### Adds a regularization term to the standard linear objective to prevent overfitting by penalizing large coefficient in linear regression equation. It useful when the dataset has multicollinearity where predictor variables are highly correlated.

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python

lr = LinearRegression()
lr.fit(X,y)
print(f"(Slope)Coefficient :{lr.coef_}")
print(f"Intercept :{lr.intercept_}")
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=10)
rr.fit(X,y)
print(f"Slope :{rr.coef_}")
print(f"Intercept :{rr.intercept_}")
```

```python
rr1 = Ridge(alpha=100)
rr1.fit(X,y)
print(f"Slope :{rr1.coef_}")
print(f"Intercept :{rr1.intercept_}")
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,rr.predict(X),color='green',label='alpha=10')
plt.plot(X,rr1.predict(X),color='orange',label='alpha=100')
plt.legend()
```

#### Here,
* At alpha = 0 the slope of Linear regression But there may be possibility having lesser slope and good Optimization
* At alpha =10 the slope is less than slope of linear regression (alpha =0)
* At alpha = 100 the slope is very less is gradually going towards UnderFitting
